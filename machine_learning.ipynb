# Conecting to housing_data_db

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine
import numpy as np
import pandas as pd

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
# Create engine using database file
engine = create_engine("sqlite:///housing_data.sqlite")
```

```python
# Declare a Base using `automap_base()`
Base = automap_base()
```

```python
# Use the Base class to reflect the database tables
Base.prepare(engine, reflect=True)
```

```python
# Print all of the classes mapped to the Base
Base.classes.keys()
```

```python
# Assign the Target class to a variable called 'target' and features to feature
target = Base.classes.Target
features = Base.classes.Features
```

```python
# Create a session
session = Session(engine)
```

```python
# Query all sales prices
results = session.query(target.index, target.sale_prc).all()
```

```python
#convert to list, remove first row, which displays the column headers
prices = list(results)
prices.pop(0)
prices
```

```python
target_df = pd.DataFrame(prices, columns=['index','SALE_PRC']).set_index('index')
```

```python
target_df
```

```python
results = session.query(features.index, features.land_sqfoot, features.tot_living_area, features.spec_feat_val, features.rail_dist, features.ocean_dist, features.water_dist, features.cntr_dist, features.subcntr_dist, features.hwy_dist, features.structure_age, features.avno60plus, features.month_sold, features.structure_quality, features.latitude_group, features.longitude_group).all()
```

```python
feature_list = list(results)
columns_list = feature_list[0]
feature_list.pop(0)
feature_list
```

```python
feature_df = pd.DataFrame(feature_list, columns=columns_list).set_index('index')
```

```python
feature_df
```

# Machine Learning

```python
# feature_df is X, target_df is y
X = feature_df
y = target_df

# Create a linear model
model = LinearRegression()
```

```python
# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Fit (train) our model to the training data
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
# Calculate feature importance in the Random Forest model.
clf = RandomForestClassifier(random_state=1, n_estimators=50).fit(X_train, y_train)
print(f'Training Score: {clf.score(X_train, y_train)}')
print(f'Testing Score: {clf.score(X_test, y_test)}')
```

# Feature Importance

```python
feature_importances = clf.feature_importances_
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
features = sorted(zip(X.columns, clf.feature_importances_), key = lambda x: x[1])
cols = [f[0] for f in features]
width = [f[1] for f in features]

fig, ax = plt.subplots()

fig.set_size_inches(10,10)
plt.margins(y=0.001)
ax.set_title('Features by level of Importance')
ax.set_xlabel('Importance')
ax.set_ylabel('Feature')

ax.barh(y=cols, width=width)

plt.savefig('feature_import.png')

plt.show()
```

# Predicted Prices vs Actual Prices

```python
# Price predictions
y_pred = model.predict(X_test)
y_pred
```

```python
X_test['predicted_price'] = y_pred
```

```python
comparison_pd = pd.merge(X_test, y_test, on="index")
comparison_pd = comparison_pd[['predicted_price','SALE_PRC']]
```

```python
comparison_pd = comparison_pd.rename(columns={'predicted_price':'Predicted Price',
                              'SALE_PRC':'Actual Price'})
comparison_pd
```

```python
comparison_pd.plot(kind="scatter", x='Predicted Price', y='Actual Price', title=("Predicted vs Actual Prices"), figsize=(10, 10))
```

# Dropping less Important Features

```python
# Create a new linear model
newModel = LinearRegression()
```

```python
# feature_df is X, target_df is y
# Drop  longitude_group, latitude_group, structure_quality, and avno60plus from feature set
X = feature_df.drop(columns=['longitude_group', 'latitude_group', 'structure_quality', 'avno60plus'])
y = target_df

# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Fit (train) our model to the training data
newModel.fit(X_train, y_train)
```

```python
newModel.score(X_test, y_test)
```

```python
# Calculate feature importance in the Random Forest model.
newCLF = RandomForestClassifier(random_state=1, n_estimators=50).fit(X_train, y_train)
print(f'Training Score: {newCLF.score(X_train, y_train)}')
print(f'Testing Score: {newCLF.score(X_test, y_test)}')
```

```python
# Price predictions
new_y_pred = newModel.predict(X_test)
new_y_pred
```

```python
X_test['predicted_price'] = new_y_pred
```

```python
new_comparison_pd = pd.merge(X_test, y_test, on="index")
new_comparison_pd = new_comparison_pd[['predicted_price','SALE_PRC']]
```
