```python
# Import Dependencies
import pandas as pd
import numpy
```

```python
# Load in file
file = "miami-housing.csv"
```

```python
df = pd.read_csv(file)
df.head(10)
```

```python
df.dtypes
```

```python
# Parcel number is the unique identifier for the land - dropping as not useful
df = df.drop(["PARCELNO"], axis='columns')
df.head(10)
```

```python
# The dataset says avno60plus is a dummy variable for airplane noise exceeding an acceptable level
df['avno60plus'].describe()
```

```python
# The column has 1 if airplane noise exceeds an acceptable level
df['avno60plus'].value_counts()
```

```python
# Long and Lat should be binned - checking the LATITUDE spread
df['LATITUDE'].describe()
```

```python
# Create 8 bins, spaced evenly
# get maximum lat, min lat
max_lat = df['LATITUDE'].max()
min_lat = df['LATITUDE'].min()
# get the difference and divide by the number of bins
lat_spacing = (max_lat - min_lat)/8
lat_bins = [1,2,3,4,5,6,7,8]

#build bins
bins = []
for x in numpy.arange(min_lat, max_lat, lat_spacing):
    bins.append(x)

bins.append(max_lat)
bins
```

```python
df["latitude_group"] = pd.cut(df["LATITUDE"], bins, labels=lat_bins)
df["latitude_group"].value_counts()
```

```python
# Long and Lat should be binned - checking the LONGITUDE spread
df['LONGITUDE'].describe()
```

```python
# Create 8 bins, spaced evenly
# get maximum long, min long
max_long = df['LONGITUDE'].max()
min_long = df['LONGITUDE'].min()
# get the difference and divide by the number of bins
long_spacing = (max_long - min_long)/8
long_bins = [1,2,3,4,5,6,7,8]

#build bins
bins = []
for x in numpy.arange(min_long, max_long, long_spacing):
    bins.append(x)

bins.append(max_long)
bins
```

```python
df["longitude_group"] = pd.cut(df["LONGITUDE"], bins, labels=long_bins)
df["longitude_group"].value_counts()
```
